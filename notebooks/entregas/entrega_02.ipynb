# **Introdução**

O desenvolvimento de energia renovável no Brasil vem se tornando cada vez mais necessário no contexto atual, impulsionado pelo aumento da população, urbanização, expansão tecnológica, industrial e comercial, além das mudanças climáticas provocadas pelos gases poluentes. Por exemplo, fontes renováveis: hídrica (energia da água dos rios), solar (energia do sol), eólica (energia do vento), biomassa (energia de matéria orgânica), geotérmica (energia do interior da Terra) e oceânica (energia das marés e das ondas) (EPE, 2022).

Além de permitir uma distribuição mais eficiente de recursos e uma redução nos custos de geração e transmissão de energia e a mitigação das mudanças climáticas.

# **Referencial Teórico**

No contexto da pesquisa acadêmica, o desenvolvimento de uma estrutura teórica abrangente é um componente crítico no estabelecimento da base para uma investigação robusta. Uma estrutura teórica serve para alinhar os objetivos da pesquisa com conceitos relevantes, que se integram na análise dos dados.

ARIMA (AutoRegressive Integrated Moving Average): Modelos ARIMA são amplamente utilizados para a previsão de séries temporais e se baseiam na ideia de que valores passados da série podem ser usados para prever valores futuros. Eles são compostos por três componentes principais:
-	Autoregressivo (AR): Utiliza valores passados da própria série temporal como variáveis preditoras.
-	Integrado (I): Refere-se à ordem de diferenciação aplicada à série temporal para torná-la estacionária, ou seja, remover tendências e sazonalidades.
-	Médias Móveis (MA): Incorpora os erros de previsão passados no modelo.

SARIMA (Seasonal AutoRegressive Integrated Moving Average): Modelos SARIMA são uma extensão dos modelos ARIMA que levam em consideração a sazonalidade presente nos dados. Eles incorporam parâmetros adicionais para modelar padrões sazonais, o que pode melhorar a precisão das previsões, especialmente em séries temporais com padrões sazonais fortes.

O primeiro trabalho analisado se insere na análise da previsão de demanda de energia elétrica, um aspecto crucial para garantir o fornecimento estável e confiável de energia. A utilização de dados reais de uma subestação e a busca pela otimização da rede LSTM através de testes iterativos fornecem insights valiosos para a definição da metodologia deste trabalho.

## Pré-processamento:
O pré-processamento é a etapa inicial e crucial em qualquer projeto de análise de dados ou aprendizado de máquina. Ele consiste em adequar os dados brutos para que estejam prontos para serem utilizados na fase de modelagem.

No contexto do nosso projeto, o pré-processamento será fundamental para garantir a qualidade dos dados e prepará-los para os modelos de séries temporais e de aprendizado de máquina que utilizaremos. 
•	Objetivo: Adequar os dados brutos à análise, garantindo qualidade e formato adequado para modelagem. As três bases de dados serão pré-processadas individualmente antes da combinação.

### Base 1: Consumo Mensal de Energia Elétrica
- Limpeza de Dados:
    - Tratamento de Valores Faltantes: Identificar e tratar os valores ausentes nas bases de dados. Como exemplo, utilizaremos a média para colunas numéricas como "Consumo" e "Consumidores" e a moda para colunas categóricas como "Região", "Sistema" e "Classe".
    
    - Tratamento de Outliers: Detectar e tratar valores discrepantes que podem distorcer as análises. Utilizaremos boxplots para visualizar e analisar a presença de outliers na coluna "Consumo", avaliando a necessidade de remoção ou tratamento.
    
    - Transformação de Dados: Os dados serão transformados para um formato adequado para análise, o que pode incluir:
        - Formato de Data: Padronizar o formato das colunas de data para o tipo datetime, o que permitirá realizar análises temporais.
        - Faremos isso para as colunas "Data" e "DataExcel".
    - Criação de Variáveis Dummy: Converter variáveis categóricas, como "Região", "Sistema", "Classe" e "TipoConsumidor", em variáveis dummy (0 ou 1), tornando-as adequadas para utilização em modelos de aprendizado de máquina.
- Agregação de Dados:
    - Agregação Temporal: Calcular valores agregados em diferentes níveis de tempo, como o consumo total mensal de energia por Unidade Federativa (UF) e subsistema.
    - Agregação por Categoria: Calcular valores agregados para diferentes categorias, como a quantidade total de consumidores, subsistema e classe. Criar novas variáveis a partir das existentes, como indicadores de sazonalidade ou tendências.

O objetivo principal do pré-processamento é entregar dados limpos, consistentes e no formato ideal para que os modelos de aprendizado de máquina consigam aprender os padrões e gerar previsões precisas. Nesse projeto, um bom pré-processamento será essencial para construirmos um modelo robusto e confiável para a projeção da demanda de energia e para gerarmos recomendações eficazes de fontes renováveis.

## **Modelagem Preditiva** 
Para lidar com a natureza temporal dos dados de consumo energético, o modelo proposto é uma Rede Neural Recorrente (RNN) com arquitetura Long Short-Term Memory (LSTM)13. A LSTM é capaz de aprender padrões complexos em sequências de dados, tornando-a adequada para prever a demanda futura de energia, levando em consideração sua variação ao longo do tempo.

- Modelos Possíveis:
    - Séries Temporais: ARIMA, SARIMA, Prophet. Capturam a natureza temporal da demanda de energia. 
    - Machine Learning: Regressão Linear, Regressão Polinomial, Árvores de Decisão.
    - Redes Neurais Artificiais (RNAs): RNNs. Capturam relações complexas e não lineares nos dados.

- Fase da modelagem: 
    - Treinamento: O modelo LSTM será treinado com os dados históricos de consumo de energia, juntamente com outras variáveis relevantes, como clima, regiões e indicadores econômicos.
    - Validação: O desempenho do modelo será avaliado com dados não utilizados no treinamento. Métricas de erro como RMSE (Root Mean Squared Error) e MAE (Mean Absolute Error) serão utilizadas para determinar a precisão das previsões.
    - Previsão de Demanda: O modelo treinado será utilizado para prever a demanda futura de energia para diferentes regiões e subsistemas. A análise temporal será aplicada para identificar padrões de consumo, como aumento da demanda em meses mais quentes. Os dados de consumo mensal por região e classe serão utilizados para treinar o modelo, considerando a sazonalidade e as tendências de consumo.  

O sistema utilizará as previsões de demanda, juntamente com informações sobre a capacidade de geração de diferentes fontes renováveis, para gerar recomendações personalizadas para cada distribuidora, região e subsistema, considerando:

-	Matriz Energética Regional: O sistema considerará as características geográficas e climáticas de cada região para recomendar as fontes de energia renovável mais adequadas. Por exemplo: eólica no Nordeste, solar em regiões com alta incidência solar, hídrica em locais com grande potencial hídrico, biomassa em regiões com alta produção agrícola.
-	Demanda Projetada: As recomendações serão baseadas na demanda de energia prevista para cada região e subsistema, garantindo o atendimento da demanda futura.
-	Sustentabilidade e Impacto Ambiental: O sistema priorizará fontes de energia com menor impacto ambiental, visando reduzir as emissões de gases de efeito estufa (GEE), alinhado com os Objetivos de Desenvolvimento Sustentável (ODS).

### **Métricas de Avaliação:**
A eficácia do sistema será avaliada por meio das seguintes métricas:

1. Precisão da Previsão da Demanda:

- RMSE, MAE e MAPE (Mean Absolute Percentage Error).

2. Eficiência da Recomendação:

- Redução de Emissões de GEE: Comparar as emissões projetadas com base nas recomendações do sistema com um cenário base, utilizando dados históricos de emissões e o fator de emissão de cada fonte de energia.

- Custos de Geração: Estimar os custos de geração de energia considerando os custos de instalação, operação e manutenção de cada fonte renovável, comparando com os custos de geração de energia a partir de fontes fósseis.

- Segurança Energética: Avaliar a resiliência do sistema energético proposto a fatores climáticos como secas, considerando a diversificação da matriz energética e a capacidade de geração de cada fonte ao longo do ano, utilizando indicadores como índice de disponibilidade e capacidade de armazenamento.
